```python
base_davvnd = base['DVI_p.CN'] #Base de Davivienda

base_davvnd['SESSION'] = base_davvnd.index.date
```

```python
base_corfi = base_corfi[['SESSION', 'EVENT_TYPE', 'SOURCE_DATETIME', 'TRDPRC_1', 'TRDVOL_1', 'BID', 'BIDSIZE', 'ASK', 'ASKSIZE', 'ACVOL_UNS', 'VWAP', 'TRNOVR_UNS']]
```

```python
base_davvnd = base_davvnd[['SESSION','EVENT_TYPE', 'SOURCE_DATETIME', 'TRDPRC_1', 'TRDVOL_1', 'BID', 'BIDSIZE', 'ASK', 'ASKSIZE', 'ACVOL_UNS', 'VWAP', 'TRNOVR_UNS']]
```

```python
lista_bases = [base_corfi, base_davvnd]
```

```python
""" Columnas que deben rellenarse solo con ffill diario
cols_ffill_por_dia = ['VWAP', 'ACVOL_UNS', 'TRNOVR_UNS']

Columnas restantes que se rellenan con 0 directamente
cols_fill_0 = [col for col in base_corfi.columns if col not in cols_ffill_por_dia]

Rellenar con 0 las columnas no acumulativas
base_corfi[cols_fill_0] = base_corfi[cols_fill_0].fillna(0)

Aplicar ffill diario a columnas acumulativas
for col in cols_ffill_por_dia:
    base_corfi[col] = (
        base_corfi.groupby(base_corfi.index.date)[col]
        .transform(lambda x: x.ffill())
        .fillna(0)
    )"""
```

```python
for i, df in enumerate(lista_bases):
  df = df.copy()
  df.loc[:,'SPREAD'] = df['ASK'] - df['BID']
  lista_bases[i] = df
```

```python
for i, df in enumerate(lista_bases):
  df.copy()
  df = df[(df.SPREAD > 0)]
  lista_bases[i] = df
```

```python
for i, df in enumerate(lista_bases):
    df = df.copy()  # recomendado
    df.loc[:, "SPRDTCK"] = pd.to_numeric(df["SPREAD"], errors="coerce").div(20)
    lista_bases[i] = df
```

```python
for i, df in enumerate(lista_bases):
  df = df.copy()
  df['IMBALANCE'] = df['BIDSIZE']/(df['ASKSIZE'] + df['BIDSIZE'])
  lista_bases[i] = df
```

```python
for i, df in enumerate(lista_bases):
  df = df.copy()
  df = df.dropna(subset = 'EVENT_TYPE')
  lista_bases[i] = df
```

```python
for i, df in enumerate(lista_bases):
  df = df.copy()
  df['BIDASK_DISTANCE'] = df['SPREAD']/2
  lista_bases[i] = df
```

```python
for i, df in enumerate(lista_bases):
  df = df.copy()
  df["INCREASE_BID"] = (
      df.groupby("SESSION", group_keys=False)
      .apply(
          lambda df: (
              # 1️⃣ Aumento del tamaño cuando el precio BID se mantiene
              (df["BIDSIZE"].diff().gt(0) & (df["BID"] == df["BID"].shift(1)))
              # 2️⃣ O nuevo mejor BID (sube el precio)
              | (df["BID"].gt(df["BID"].shift(1)))
          ).cumsum(),
          include_groups=False
      )
  )

  df["INCREASE_BIDSIZE"] = (
      df.assign(
          bid_diff = df.groupby("SESSION")["BIDSIZE"].diff(),
          price_diff = df.groupby("SESSION")["BID"].diff()
      )
      .pipe(lambda df: np.where(
          # 1️⃣ Aumento del tamaño con mismo precio
          (df["bid_diff"] > 0) & (df["price_diff"] == 0), df["bid_diff"],
          # 2️⃣ Nuevo mejor BID (precio sube → todo el tamaño cuenta)
          np.where(df["price_diff"] > 0, df["BIDSIZE"], 0)
      ))
  )

  df['INCREASE_AC_BIDSIZE'] = df['INCREASE_BIDSIZE'].cumsum()

  lista_bases[i] = df
```

```python
for i, df in enumerate(lista_bases):

  df = df.copy()
  df["INCREASE_ASK"] = (
      df.groupby("SESSION", group_keys=False)
      .apply(
          lambda df: (
              # 1️⃣ Aumento del tamaño cuando el precio ASK se mantiene
              (df["ASKSIZE"].diff().lt(0) & (df["ASK"] == df["ASK"].shift(1)))
              # 2️⃣ O nuevo mejor ASK (baja el precio)
              | (df["ASK"].lt(df["ASK"].shift(1)))
          ).cumsum(),
          include_groups=False
      )
  )

  df["INCREASE_ASKSIZE"] = (
      df.assign(
          ASK_diff = df.groupby("SESSION")["ASKSIZE"].diff(),
          price_diff = df.groupby("SESSION")["ASK"].diff()
      )
      .pipe(lambda df: np.where(
          # 1️⃣ Aumento del tamaño con mismo precio
          (df["ASK_diff"] > 0) & (df["price_diff"] == 0), df["ASK_diff"],
          # 2️⃣ Nuevo mejor ASK (precio baja → todo el tamaño cuenta)
          np.where(df["price_diff"] < 0, df["ASKSIZE"], 0)
      ))
  )

  df['INCREASE_AC_ASKSIZE'] = df['INCREASE_ASKSIZE'].cumsum()

  lista_bases[i] = df
```

```python
for i, df in enumerate(lista_bases):
  df = df.copy()


  # --- REDUCTION_BID (contador de eventos)
  mask_reduction_bid = (
      (df["BIDSIZE"].diff().lt(0) & (df["BID"] == df["BID"].shift(1))) |
      (df["BID"].lt(df["BID"].shift(1)))
  )
  df["REDUCTION_BID"] = mask_reduction_bid.groupby(df["SESSION"]).cumsum()

  # --- REDUCTION_BIDSIZE (volumen puntual retirado)
  df["REDUCTION_BIDSIZE"] = (
      df.assign(
          bid_diff = df.groupby("SESSION")["BIDSIZE"].diff(),
          price_diff = df.groupby("SESSION")["BID"].diff()
      )
      .pipe(lambda df: np.where(
          # 1️⃣ Disminuye tamaño con mismo precio
          (df["bid_diff"] < 0) & (df["price_diff"] == 0), -df["bid_diff"],
          # 2️⃣ Nuevo peor BID (precio baja → todo el tamaño previo se retira)
          np.where(df["price_diff"] < 0, df["BIDSIZE"].shift(1), 0)
      ))
  )

  # --- REDUCTION_AC_BIDSIZE (volumen acumulado por sesión)
  df["REDUCTION_AC_BIDSIZE"] = (
      df.groupby("SESSION")["REDUCTION_BIDSIZE"].cumsum()
  )

  lista_bases[i] = df
```

```python
for i, df in enumerate(lista_bases):
  df = df.copy()

  # --- REDUCTION_ASK (contador de eventos)
  mask_reduction_ask = (
      (df["ASKSIZE"].diff().lt(0) & (df["ASK"] == df["ASK"].shift(1))) |
      (df["ASK"].gt(df["ASK"].shift(1)))
  )
  df["REDUCTION_ASK"] = mask_reduction_ask.groupby(df["SESSION"]).cumsum()

  # --- REDUCTION_ASKSIZE (volumen puntual retirado)
  df["REDUCTION_ASKSIZE"] = (
      df.assign(
          ASK_diff = df.groupby("SESSION")["ASKSIZE"].diff(),
          price_diff = df.groupby("SESSION")["ASK"].diff()
      )
      .pipe(lambda df: np.where(
          # 1️⃣ Disminuye tamaño con mismo precio
          (df["ASK_diff"] < 0) & (df["price_diff"] == 0), -df["ASK_diff"],
          # 2️⃣ Nuevo peor ASK (precio sube → todo el tamaño previo cuenta)
          np.where(df["price_diff"] > 0, df["ASKSIZE"].shift(1), 0)
      ))
  )

  # --- REDUCTION_AC_ASKSIZE (volumen acumulado por sesión)
  df["REDUCTION_AC_ASKSIZE"] = (
      df.groupby("SESSION")["REDUCTION_ASKSIZE"].cumsum()
  )

  lista_bases[i] = df
```

```python
for i, df in enumerate(lista_bases):
  df = df.copy()

  # 1️⃣ Cambios netos por lado
  df["DELTA_B"] = df["INCREASE_BIDSIZE"] - df["REDUCTION_BIDSIZE"]
  df["DELTA_A"] = df["INCREASE_ASKSIZE"] - df["REDUCTION_ASKSIZE"]

  # 2️⃣ Order Flow Imbalance puntual
  df["OFI"] = df["DELTA_B"] - df["DELTA_A"]

  # 3️⃣ Acumulado por sesión
  df["OFI_CUM"] = df.groupby("SESSION")["OFI"].cumsum()

  lista_bases[i] = df
```
